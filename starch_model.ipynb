```python
#11-01 モジュール・パッケージの読み込み
import numpy as np
import matplotlib.pyplot as plt
```

```python
#11-02 パラメータ・初期値
a = 140
gamma = 0.6
H = 10
tL = 16/24
TL = 12/24
Cdawn = 10
C0 = 10
S0 = 8
```

```python
#11-03 デンプン代謝モデル
dt = 0.01
t = 0
C = C0
S = S0
t_list = []
C_list = []
S_list = []
beta_list = []

for i in range(100):
  t = dt*(i+1)
  if t <= TL:
    Lt = 1
  else:
    Lt = 0
  betaL = a*(gamma-1+tL)/(a*(1-tL)*t+Cdawn)**(2/3)
  betaD = a*tL/(a*tL*(1-t)+Cdawn)**(2/3)
  beta = Lt*betaL + (1-Lt)*betaD
  dCdt = a*gamma*Lt - beta*C**(2/3)
  dSdt = a*(1-gamma)*Lt + beta*C**(2/3) - H*S
  C = C + dCdt*dt
  S = S + dSdt*dt
  t_list.append(t)
  C_list.append(C)
  S_list.append(S)
  beta_list.append(beta)
```

```python
#11-04 プロット
plt.plot(t_list,C_list,label="Starch")
plt.plot(t_list,S_list,label="Sucrose")
plt.plot(t_list,beta_list,label="Degradation")
plt.legend()
plt.show()
```
